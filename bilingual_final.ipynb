```python
data = pd.read_csv("../data/data_w_new_lang.csv")
```

```python
data_1 = data.loc[(data['proj_year'] > 2009) & (data['proj_year'] < 2016) &
                (data['proj_class'] == '面上项目') &
                (data['ptype'].isin(['journal', 'conference']))]
data_2 = data.loc[(data['proj_year'] > 2009) & (data['proj_year'] < 2016) &
                (data['proj_class'] == "青年科学基金项目") &
                (data['ptype'].isin(['journal', 'conference']))]
data_3 = data.loc[(data['proj_year'] > 2009) & (data['proj_year'] < 2015) &
                (data['proj_class'] == '重点项目') &
                (data['ptype'].isin(['journal', 'conference']))]
data_4 = data.loc[(data['proj_year'] > 2009) & (data['proj_year'] < 2016) &
                (data['proj_class'] == '地区科学基金项目') &
                (data['ptype'].isin(['journal', 'conference']))]
data = pd.concat([data_1, data_2, data_3, data_4])
```

```python
data['is_Chinese_paper_lingua'] = data['title'].map(lambda a: evaluate_language_lingua(a))

data['is_Chinese_paper_lingua'].value_counts()
```

```python
data = data.loc[(data.year_processed > 2009) & (data.year_processed < 2021)]
```

```python
summary_language_count = data.is_Chinese_paper_lingua.value_counts().reset_index()
summary_language_count.columns = ['language', 'count']
language_list = summary_language_count.loc[summary_language_count['count'] > 1000, 'language']
print(language_list)
```

```python
data_final = data.loc[data.is_Chinese_paper_lingua.isin(language_list)]
```

```python
def language_redefine(x):
    if x == "Language.CHINESE":
        return("Chinese")
    else:
        return("English")

data_final['is_Chinese_paper_lingua_f'] = data_final.is_Chinese_paper_lingua.map(lambda x: language_redefine(x))
```
